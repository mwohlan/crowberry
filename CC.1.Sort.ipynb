### Filter HQ Country to Nordics

```python
countries = df['HQ COUNTRY'].unique()
nordics = ['Sweden', 'Norway', 'Iceland', 'Denmark', 'Finland']
df = df[df['HQ COUNTRY'].isin(nordics)]

df.reset_index()
```

### Delete unwanted growth stages:

```python
neglected_growth_stages = ['late growth stage', 'not meaningful']

indexrows = df[df['GROWTH STAGE'].isin(neglected_growth_stages)].index

df = df.drop(indexrows)
df.reset_index()
df.shape
```

### Neglect certain GROWTH STAGE

```python
neglected_last_rounds = ['ACQUISITION',
       'LATE VC', 'IPO', 'MERGER', 'POST IPO DEBT', 'POST IPO EQUITY',
       'GROWTH EQUITY NON VC', 'GROWTH EQUITY VC',
       'POST IPO CONVERTIBLE', 'SERIES C', 'SERIES B', 'SERIES D']

indexrowsLR = df[df['LAST ROUND'].isin(neglected_last_rounds)].index

df = df.drop(indexrowsLR)
df.reset_index()
df.shape
```

### Neglect LAST ROUND before 01/2018

```python
# Sort out for last funding dates older than jan 2018
indexLFD = df[df['LAST FUNDING DATE'] < '2018-01-01'].index

df = df.drop(indexLFD)
df.reset_index()
df.shape

#df['LAST FUNDING DATE'].unique()
#Reduces to 4086
```

### Neglect zero value records

```python
#Check if LAST FUNDING DATE, FIRST FUNDING DATE, SEED YEAR, REVENUE AS LIST, VALUATION DATE IS ZERO, FOUNDERS GENDER

index_of_null_records = df[(df['LAST FUNDING DATE'].isnull() & df['FIRST FUNDING DATE'].isnull() & df['SEED YEAR'].isnull()& df['VALUATION DATE'].isnull()  & df['FOUNDERS GENDERS'].isnull())].index

df = df.drop(index_of_null_records)
df.reset_index()
```

```python
indexMT = df[df['FEMALE FOUNDER'] == 0].index

df = df.drop(indexMT)
df.reset_index()
```
